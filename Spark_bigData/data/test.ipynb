# 문제1

```python
myrdd=spark.sparkContext.textFile(os.path.join("data","20191021_policeAddress.txt"))
```

```python
myrdd.collect()
```

```python
# , . '' >삭제해주기
_myrdd=myrdd.map(lambda x:x.replace("‘"," ").replace("’"," ").replace("·"," ")).flatMap(lambda x:x.split( ))
_myrdd.collect()
```

```python
#한글자짜리 제외해주기
_myrdd=_myrdd.filter(lambda x:len(x)!=1)
_myrdd.collect()
```

```python
#빈도세기
bin=_myrdd.map(lambda x:(x,1))
bin.collect()
```

```python
#빈도수 세고 빈도순으로 (내림차순)으로 출력
_bin=bin.groupByKey().mapValues(sum).map(lambda x:(x[1],x[0])).sortByKey(False)
_bin.collect()
```

# 문제2

```python
rdd2=spark.sparkContext.textFile("data/BUS_STATION_BOARDING_MONTH_201909_small.csv")
```

```python
_rdd2=rdd2.map(lambda x:x.split(','))
_rdd2.collect()
```

```python
#문제 1컬럼추출
_rdd2.map(lambda x:(x[2],x[8],x[9])).collect()
```

```python
#문제 2-2
head=_rdd2.first()
_rdd2=_rdd2.filter(lambda x:x != head) #첫줄제거하기
_rdd2=_rdd2.filter(lambda x:x!=")
_rdd2.collect()
```

```python
type(_rdd2.collect()[0][8])
```

```python

sum_rdd2=_rdd2.map(lambda x:(x[2],x[8][1:3]))
sum_rdd2.collect()
```

```python
sum_rdd2=sum_rdd2.reduceByKey(lambda x,y:x+y)
sum_rdd2.collect()
for k,v in sum_rdd2.collect() :
    print("버스번호:",k ,"합계:",v)
```
